```python
# Create a "Correlational Pearson's Matrix" 
corr_mat = Ghd_dropped.corr(method = 'pearson')
sns.set_style('whitegrid')
mask = np.zeros_like(corr_mat, dtype = bool)
mask[np.triu_indices_from(mask)] = True
fig, ax = plt.subplots(figsize=(15, 15))
cmap = sns.diverging_palette(226, 15, as_cmap = True)
sns.heatmap(corr_mat, mask = mask, cmap = cmap, vmin =- 1, vmax = 1, center = 0, 
            square = True, linewidth = 0.5, cbar_kws = {'shrink': 1}, annot = True, annot_kws = {'size': 15})
plt.tight_layout()
fig.show()
```
